```python
%%R
#Instalando pacotes
install.packages('/content/expert_1_0-0.tar.gz', repos=NULL, type="source")
#carregando a lib
library(expert)
#Criando uma lista com as informações especialistas
x <- list (EXP1 <- list(INT <- c(16, 32, 190000)),
           EXP2 <- list(INT <- c(110000, 120000, 130000)),
           EXP3 <- list(INT <- c(200000, 300000, 400000)))
```

```python
%%R
prob <- c(0.1, 0.5, 0.9)

inf <- expert(x, "weights", prob, w=c(.6, .2, .2))

hist(inf, col="blue")
```

```python
%%R
print(summary(inf))

print(quantile(inf))

cat("\n")
cat("Valor médio", mean(inf))
```

```python
%%R
par(bg = "white")
split.screen(c(2,2))

screen(1)
hist(inf, col = "gray", main = "Distribuição agregada")

screen(2)
s = density(c(15000, 180000, 190000))
plot(s, main="Especialista 1")
polygon(s, col="blue")

screen(3)
s = density(c(110000, 120000, 130000))
plot(s, main="Especialista 2")
polygon(s, col="blue")

screen(4)
s = density(c(200000, 300000, 400000))
plot(s, main="Especioalista 3")
polygon(s, col="blue")
close.screen(all = TRUE)
```
